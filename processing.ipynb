```python
import numpy as np
x = []
y = []
u = []
v = []
Nx = 256
Ny = 140
with open("result.csv",'r') as f:
    curLine = f.readline()
    while True:
        curLine = f.readline()
        if curLine == '': 
            break
        data = curLine.split(',')
        x.append(float(data[1]))
        y.append(float(data[2]))
        u.append(float(data[3]))
        v.append(float(data[4]))
x = np.array(x).reshape(Nx,Ny)
y = np.array(y).reshape(Nx,Ny)
u_r = np.array(u)
v_r = np.array(v)
u = np.zeros((Nx,Ny))
v = np.zeros((Nx,Ny))
for i in range(Nx):
    for j in range(Ny):
        u[i,j] = u_r[i*Ny+j]
        v[i,j] = v_r[i*Ny+j]
#x = x[::Ny]
#y = y[:Ny]

print(x,y)
```

```python
#x = np.arange(Nx)
#y = np.arange(Ny)
import matplotlib.pyplot as plt
speed = np.sqrt(u**2+v**2).T
fig_size = 0.05
plt.figure(dpi=200,figsize=(256*fig_size,102*fig_size))
plt.streamplot(x.T,y.T,u.T,v.T,broken_streamlines=False,linewidth=8*speed/speed.max())
plt.show()
```
